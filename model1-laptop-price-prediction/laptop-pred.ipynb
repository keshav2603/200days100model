```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler

from sklearn.pipeline import make_pipeline

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
df= pd.read_csv("Laptop_price.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df["Brand"].value_counts()
```

```python
df.info()
```

```python
df["Price"] =df["Price"].round(2)
df["Weight"] =df["Weight"].round(1)
df["Processor_Speed"] = df["Processor_Speed"].round(2)
df["Screen_Size"] = df["Screen_Size"].round(1)
```

```python
df
```

```python
x=df.drop(columns=["Price"])
y= df["Price"]
```

```python
df.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train.shape
```

```python
#trf1 one hot encoidng transformer

trf1=ColumnTransformer([
    ("ohe_brand",OneHotEncoder(sparse_output=False,drop="first",handle_unknown="ignore"),[0])
],remainder="passthrough")
```

```python
#trf2 scaling all the value
trf2 = ColumnTransformer([
    ("scaling", MinMaxScaler(),slice(0,8))
])
```

```python
#third step our model

trf3 = LinearRegression()
```

```python
pipe=make_pipeline(trf1, trf2, trf3)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
# as high as possible 
r2_score(y_pred,y_test)
```

```python
# as low as possible
np.sqrt(mean_squared_error(y_pred, y_test))
```

```python
# as low as possible rmse
mean_absolute_error(y_pred, y_test)
```
